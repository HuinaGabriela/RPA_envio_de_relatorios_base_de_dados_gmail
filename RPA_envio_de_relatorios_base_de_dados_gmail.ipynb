## Desafio

- Vamos automatizar a extração de informações de um sistema e envio de um relatório por e-mail
- No nosso caso, para todo mundo conseguir fazer o mesmo programa, o nosso "sistema" vai ser o Gmail, mas o mesmo processo pode ser feito com qualquer programa do seu computador e qualquer sistema
    - Passo 1: Entrar no sistema (entrar no Gmail)
    - Passo 2: Entrar em uma aba específica do sistema onde tem o nosso relatório (Aba Contatos)
    - Passo 3: Exportar o Relatório (Exportar Contatos)
    - Passo 4: Pegar o relatório exportado, tratar e pegar as informações que queremos
    - Passo 5: Preencher/Atualizar informações do sistema (No nosso caso, criar um e-mail e enviar)


### Principais comandos

- pyautogui.write() --> escreve
- pyautogui.click --> clica
- pyautogui.LocateOnScreen --> identifica uma imagem na tela
- pyautogui.hotkey --> usa atalhos do teclado, combinação de teclas do teclado
- pyautogui.press --> aperta um botão do teclado

- abrir o sistema (abrir o navegador - Google Chrome)

- apertar o windows do teclado
    - print(pyautogui.KEYBOARD_KEYS)
- digitar o chrome
- apertar enter
- entrar no gmail



```python
import pandas as pd
import pyperclip
import pyautogui
import time

pyautogui.alert('O código vai começar! NÃO MEXA NO COMPUTADOR E MOUSE enquanto estiver executando o script!. Quando finalizar, eu te aviso')

pyautogui.PAUSE = 1
# - abrir o sistema (abrir o navegador - Google Chrome)

# apertar o windows do teclado
pyautogui.press('win')
# digitar o chrome
pyautogui.write('chrome')
# apertar enter
pyautogui.press('enter')
time.sleep(3)

# procurar contatos e acessar os contatos
x, y, largura, altura = pyautogui.locateOnScreen('pontinhos_pontinhos.png')
time.sleep(3)
pyautogui.click(x + largura/2, y + altura/2)
x, y, largura, altura = pyautogui.locateOnScreen('contatos_.png')
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(3)

# mover o mouse
# print(pyautogui.position())
# Point(x=768, y=338)
pyautogui.moveTo(800, 150, duration=2)
time.sleep(2)

# exportar contatos pessoal e confirmar
x, y, largura, altura = pyautogui.locateOnScreen('grupo_pessoal.png')
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(2)

x, y, largura, altura = pyautogui.locateOnScreen('exportar_grupo_pessoal.png')
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(2)

x, y, largura, altura = pyautogui.locateOnScreen('exportar_grupo_pessoal_confirmar.png')
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(4)

#################################
# fazer download da base de dados
#################################

df = pd.read_csv(r'C:/Users/Administrador/Downloads/contacts.csv')
#df = df.dropna(axis=1)# exclui colunas vazias
display(df)
time.sleep(4)

# vai ter que ir nos pontinhos
x, y, largura, altura = pyautogui.locateOnScreen('pontinhos_do_contato.png')
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(3)
# acessar o gmail
x, y, largura, altura = pyautogui.locateOnScreen('gmail_contatos.png')
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(2)

# print(pyautogui.position())
# Point(x=768, y=338)
pyautogui.moveTo(200, 100, duration=3)
time.sleep(3)

# # esperar contatos carregar(bug ao reconhecer imagem)
while not pyautogui.locateOnScreen('carregar_gmail.png'):
    time.sleep(4)

# clicar em escrever email ## TESTEI ATE AQUIII!!!##

for email in df['E-mail 1 - Value']:#caso queira mandar para todos
    # clicar no botao escrever
    x, y, largura, altura = pyautogui.locateOnScreen('escreverEmail.png')
    pyautogui.click(x + largura/2, y + altura/2)

    #escrever o email
    pyautogui.write(email)
    # enter
    pyautogui.press('enter')
    #tab muda direto para assunto senão funcionar usa imagem mesmo do assunto
    pyautogui.press('tab')
    #assunto do email
    pyautogui.write('teste rpa huina')
    time.sleep(3)
    #tab para o corpo do email
    pyautogui.press('tab')
    time.sleep(3)
    # tem que tratar as acentuações aff
    #pyperclip ajuda a dar ctrl+C , podemos fazer assim
    texto = '''
        Bom dia,
        Estou fazendo um teste de rpa,
        meu texto precisa mostrar as acentuações
        corretamente (posição, à, índice)
        Abraços. '''

    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')
    # enviar -->> usar atalho do gmail
    pyautogui.hotkey('ctrl', 'enter')
pyautogui.alert('Script finalizado! Pode pegar seu computador de volta!')

```
